```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
solicitacao_coleta = {
    'SUPERINTENDENCIA': ['LRB', 'LRC', 'LRN', 'LRO', 'LRS'],
    'Jan': [42, 44, 39, 55, 20],
    'Fev': [32, 57, 55, 60, 19],
    'Mar': [40, 58, 27, 32, 25]
}

df = pd.DataFrame(solicitacao_coleta)
df
```

```python
# Gráfico de barras
plt.figure(figsize=(5,3))
plt.barh(df['SUPERINTENDENCIA'], df['Jan'])
plt.title('Solicitações de Coleta do Mês Janeiro/26')
plt.xlabel('Quant.')
plt.ylabel('Super.')
plt.show()
```

```python
# Gráfico de Linhas
plt.figure(figsize=(5,3))
plt.plot(df['SUPERINTENDENCIA'], df['Fev'], marker='o')
plt.title('Solicitações de Coleta do Mês Fevereiro/26')
plt.xlabel('Super.')
plt.ylabel('Quant.')
plt.grid(True)
plt.show()
```

```python
# Gráfico de colunas
plt.figure(figsize=(5,3))
plt.bar(df['SUPERINTENDENCIA'], df['Mar'])
plt.title('Solicitações de Coleta do Mês Março/26')
plt.xlabel('Quant.')
plt.ylabel('Super.')
plt.show()
```

```python
solicitacoes_total_janeiro = df['Jan'].sum()
print('Solicitações de coleta da DLU: ', solicitacoes_total_janeiro)
```

```python
solicitacoes_total_fevereiro = df['Fev'].sum()
print('Solicitações de coleta da DLU: ', solicitacoes_total_fevereiro)
```

```python
solicitacoes_total_marco = df['Mar'].sum()
print('Solicitações de coleta da DLU: ', solicitacoes_total_marco)
```

## Conclusão
Ao observar os gráficos de cada mês observa-se que as Super. LRO e a LRN obtiveram grandes número de solicitações de Coleta da Ouvidoria em Janeiro e Fevereiro, já em Março as dimuições foram perceptíveis. A Super. LRB esta sempre crescendo ou dimuindo, não existindo um padrão a ser observado AINDA. A Super. LRS possue baixas solicitações nos três meses do ano, o que é um ponto positivo para as gerências. Já a Super. LRC esta sempre em alta nos três meses do ano, o que é algo a se analisar por ser um ponto negativo.
